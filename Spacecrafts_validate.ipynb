```python
# веса которые мы отправили на соревнования
import time
from ultralytics import YOLO

model = YOLO('/workspace/competitions/Spacecrafts/best1.pt')

start_time = time.time()

results = model.val(data='data.yaml',device = 'cpu')

end_time = time.time()

validation_time = end_time - start_time

print(results)
print(f"Время, затраченное на валидацию: {validation_time:.2f} секунд")
```

```python
import time
from ultralytics import YOLO

# Загрузка модели YOLO
model = YOLO('/workspace/competitions/Spacecrafts/best1.pt')

# Замер времени начала валидации
start_time = time.time()

# Выполнение валидации модели на данных, указанных в файле data.yaml
results = model.val(data='data.yaml', device='cpu')

# Замер времени окончания валидации
end_time = time.time()

# Подсчет времени, затраченного на валидацию
validation_time = end_time - start_time

# Вывод времени валидации
print(f"Время, затраченное на валидацию: {validation_time:.2f} секунд")

# Вывод всех метрик для диагностики
print("Доступные метрики:", results.metrics)

# Получение и вывод IoU score, если он доступен
if 'box' in results.metrics and hasattr(results.metrics['box'], 'iou'):
    iou_score = results.metrics['box'].iou
    print(f"IoU score: {iou_score}")
else:
    print("Метрика IoU не найдена в результатах валидации.")
```
